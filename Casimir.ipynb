# Casimir Force 

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from math import pi
R1, R2_1 = (36.1e-6,29.8e-6)
R2_2 = 38.0e-6
R2_3 = 46.9e-6
d = np.arange(100e-9,500e-9,100e-11)
h_cross = 1.054e-34
c = 3e8
```

```python
tuple = (R1,R2_1)
tuple = (R1,R2_2)
tuple = (R1,R2_3)
R_prime_1 = R1*R2_1/(R1+R2_1)
R_prime_2 = R1*R2_2/(R1+R2_2)
R_prime_3 = R1*R2_3/(R1+R2_3)
greater_indices = np.where(d > 300e-9)

lesser_indices = np.where(d <= 300e-9)
```

```python
F_pp_greater =  h_cross * c * pi**2 / (240.0*d[greater_indices]**4 )

F_pp_lesser =  h_cross * c * pi**2 / (240.0*d[lesser_indices]**4 )
```

```python
Y1 = np.zeros(np.size(d))

Y1[greater_indices] = 2 * pi * F_pp_greater * 1 * d[greater_indices] * (1 / R_prime_1)  + 2* pi * F_pp_greater
Y1[lesser_indices] = 2 * pi * F_pp_lesser * 0.6 * d[lesser_indices] * (1 / R_prime_1)  + 2* pi * F_pp_lesser

plt.title('R1 = 36.1μm, and R2 = 29.8μm')
plt.xlabel('Separation')
plt.ylabel('Force(Nm-2)')
plt.plot(d,Y1,"blue")
```

```python
Y2 = np.zeros(np.size(d))

Y2[greater_indices] = 2 * pi * F_pp_greater * 1 * d[greater_indices] * (1 / R_prime_2)  + 2* pi * F_pp_greater
Y2[lesser_indices] = 2 * pi * F_pp_lesser * 0.6 * d[lesser_indices] * (1 / R_prime_2)  + 2* pi * F_pp_lesser
plt.title('R1 = 36.1μm, and R2 = 38.0μm')
plt.xlabel('Separation')
plt.ylabel('Force(Nm-2)')
plt.plot(d,Y2,"red")
```

```python

Y3 = np.zeros(np.size(d))
Y3[greater_indices] = 2 * pi * F_pp_greater * 1 * d[greater_indices] * (1 / R_prime_3)  + 2* pi * F_pp_greater
Y3[lesser_indices] = 2 * pi * F_pp_lesser * 0.6 * d[lesser_indices] * (1 / R_prime_3)  + 2* pi * F_pp_lesser

plt.title('R1 = 36.1μm, and R2 = 46.9μm')
plt.xlabel('Separation')
plt.ylabel('Force(Nm-2)')
plt.plot(d,Y3,color = "green")
```
